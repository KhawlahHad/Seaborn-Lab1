## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))  

sns.barplot(x='sex', y='survived', data=df, ci=None) 

plt.xlabel('Sex')
plt.ylabel('Survived Count')
plt.title('Survived People by Sex')

plt.show()
```

# Q10: Draw catplot represents survived people based on embarked

```python
plt.figure(figsize=(10, 6))  

sns.catplot(x='embarked', hue='survived', kind='count', data=df, palette='Set1')

plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.title('Survived People by Port of Embarkation')

plt.show()
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
plt.figure(figsize=(10, 6))  

sns.boxplot(x='pclass', y='age', hue='sex', data=df, palette='Set2')

plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Distribution of Age by Passenger Class and Gender')

plt.show()
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
selected_columns = ['sibsp', 'parch', 'age', 'fare', 'survived']

correlation_matrix = df[selected_columns].corr()

plt.figure(figsize=(10, 8)) 

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Heatmap')
plt.show()
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
plt.figure(figsize=(8, 6))

sns.catplot(x='sibsp', hue='survived', kind='count', data=df, palette='Set2')

plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Count')
plt.title('Relationship between Siblings/Spouses and Survival')

plt.show()
```

# Q14: Draw extra insights [Optional]

```python
survival_rate_by_class = df.groupby('pclass')['survived'].mean().reset_index()

plt.figure(figsize=(8, 6)) 

sns.barplot(x='pclass', y='survived', data=survival_rate_by_class, palette='Set1')

plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')

plt.show()
```
